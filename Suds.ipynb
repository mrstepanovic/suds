## Import Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import json, requests
import bdb_utils
```

## Download Data

```python
API_KEY = "YOUR_KEY_HERE"
URL = "http://api.brewerydb.com/v2/beers/"
params = dict(key=API_KEY, withBreweries='Y')

beer = bdb_utils.download_beer_data(URL, params, 500).drop_duplicates()
beer.apply(pd.to_numeric, errors='ignore')
```

## Test for Autocorrelation between Beer Metrics

```python
metrics = []'ABV', 'IBU', 'SRM', 'OG']
pd.plotting.scatter_matrix(beer[metrics], 
                           alpha=0.2, 
                           figsize=(12, 12), 
                           diagonal='kde')
```
